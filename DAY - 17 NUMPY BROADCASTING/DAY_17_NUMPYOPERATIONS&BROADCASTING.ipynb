```python
goals_matrix = np.array([
    [2, 1, 0, 3, 1],   # Match 1
    [1, 0, 2, 1, 0],   # Match 2
    [3, 2, 1, 0, 2],   # Match 3
    [0, 1, 1, 2, 1]    # Match 4
])
print(goals_matrix.shape)
```

Add a flat home-advantage bonus of 1 goal to every single value:


```python
boosted = goals_matrix + 1
print(boosted)
```

```python
player_bonus = np.array([0, 1, 0, 0, 1])   # Zeeshan and Usman get +1 (man-of-match streak)
print(goals_matrix.shape, player_bonus.shape)

boosted_matrix = goals_matrix + player_bonus
print(boosted_matrix)
```

axis-based aggregation (sum, mean, max with axis)

```python
total_per_player = goals_matrix.sum(axis=0)
print(total_per_player)
```

```python
total_per_match = goals_matrix.sum(axis=1)
print(total_per_match)
```

```python
avg_goals_per_player = goals_matrix.mean(axis=0)
print(avg_goals_per_player)

best_match_score = goals_matrix.max(axis=1)
print(best_match_score)
```
