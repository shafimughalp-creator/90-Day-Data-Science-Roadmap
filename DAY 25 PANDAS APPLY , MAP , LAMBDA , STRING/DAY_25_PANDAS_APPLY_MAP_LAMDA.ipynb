Build price_tier with .apply(axis=1) (Premium ≥2000, Standard ≥500, Budget below).

```python
def new_price_tier(row) :
  if row['price'] >= 2000 :
    return "Premium"
  elif row['price'] >= 500 :
    return "Standard"
  else :
    return "Budget Below"
orders_df['price_category'] = orders_df.apply(new_price_tier,axis=1)
orders_df
```

Add a discount_status column: "discount" if product contains "Combo" or "Bundle", else "full price".

```python
def Buncomb(row):
    if 'Bundle' in row['product']:
        return 'Bundle'
    elif 'Combo' in row['product']:
        return 'Combo'
    else:
        return 'full price'

orders_df['Bundle_Combo'] = orders_df.apply(Buncomb, axis=1)
orders_df
```

Use groupby (Cumulative Ledger) to find average price per price_tier.

```python
tier_avg = orders_df.groupby('price_category')['price'].mean()
# groupby('price_category')  -> group rows by the Premium/Standard/Budget label
# ['price']                  -> pick just the price column to summarize
# .mean()                    -> average price within each group

tier_avg
```
