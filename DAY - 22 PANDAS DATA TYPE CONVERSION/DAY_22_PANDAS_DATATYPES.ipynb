Create a DataFrame with a "team" column of 6 values, only 2 unique team names repeated. Convert it to category and compare .memory_usage(deep=True) before/after.

You're given pd.Series(["5", "6.5", "seven", "8"]). Convert to numeric with errors='coerce'. What are the resulting dtype and values — and why is "6.5" handled differently from "5" even though both look numeric?

```python
new_mini_project =  pd.Series(["5", "6.5", "seven", "8"])
fixed_n = pd.to_numeric(new_mini_project,errors="coerce")   # seven as NAN
print(fixed_n)    # errro fixed
print(fixed_n.dtype)          # becuase 6.5 was in the series it took it and rest all the numbers got automatically converted to float like we say it implicit conversion
```
