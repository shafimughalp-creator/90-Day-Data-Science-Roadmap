# Football Dataset Practice — Days 1-7 Review
## FC Lahore Lions Season Analysis
Applying Mean/Median/Mode, Variance/Std Dev, Percentiles/IQR/Outliers, Distribution/Skewness, Probability, and Bayes' Theorem on real match data.

## Load the dataset

```
import pandas as pd
df = pd.read_csv('lions_season.csv')
df.head()
```

## Day 2: Mean, Median, Mode

```
print("Mean goals:", df['goals_scored'].mean())
print("Median goals:", df['goals_scored'].median())
print("Mode goals:", df['goals_scored'].mode().tolist())
```

## Day 3: Variance & Standard Deviation

```
print("Possession variance:", df['possession_pct'].var())
print("Possession std dev:", df['possession_pct'].std())
print("Goals std dev:", df['goals_scored'].std())
```

## Day 4: Percentiles, IQR, Outlier Detection

```
q1 = df['shots_on_target'].quantile(0.25)
q3 = df['shots_on_target'].quantile(0.75)
iqr = q3 - q1
lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
print(f"Q1={q1}, Q3={q3}, IQR={iqr}")
print(f"Normal range: {lower} to {upper}")

outliers = df[(df['shots_on_target'] < lower) | (df['shots_on_target'] > upper)]
outliers[['match_no','opponent','shots_on_target','goals_scored']]
```

## Day 5: Distribution Shape & Skewness

```
from scipy import stats
print("Goal counts breakdown:\n", df['goals_scored'].value_counts().sort_index())
print("Skewness:", stats.skew(df['goals_scored']))
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['goals_scored'], bins=range(0,10), color='#2e7d32', edgecolor='black', align='left', rwidth=0.8)
plt.axvline(df['goals_scored'].mean(), color='red', linestyle='--', linewidth=2, label=f"Mean = {df['goals_scored'].mean():.2f}")
plt.axvline(df['goals_scored'].median(), color='blue', linestyle='--', linewidth=2, label=f"Median = {df['goals_scored'].median():.1f}")
plt.title("Lions — Goals Scored Per Match")
plt.xlabel("Goals Scored")
plt.ylabel("Number of Matches")
plt.legend()
plt.show()
```

## Day 6: Probability Basics

```
p_win = (df['result']=='W').mean()
p_home = (df['venue']=='Home').mean()
p_win_given_home = df[df['venue']=='Home']['result'].eq('W').mean()
p_win_given_away = df[df['venue']=='Away']['result'].eq('W').mean()

print(f"P(Win) = {p_win:.2f}")
print(f"P(Home) = {p_home:.2f}")
print(f"P(Win | Home) = {p_win_given_home:.2f}")
print(f"P(Win | Away) = {p_win_given_away:.2f}")
```

## Day 7: Bayes' Theorem
Reversing P(Win|Home) into P(Home|Win)

```
p_home_given_win = (p_win_given_home * p_home) / p_win
print(f"P(Home | Win) = {p_home_given_win:.2f}")
```

## Key Takeaways
- Mean can be misleading when outliers exist — median/mode tell the typical story
- Outliers are data points to investigate, not delete
- P(A|B) ≠ P(B|A) — Bayes' Theorem is required to correctly flip direction
- Conditional probabilities on different subsets (Home vs Away) are NOT complements of each other
