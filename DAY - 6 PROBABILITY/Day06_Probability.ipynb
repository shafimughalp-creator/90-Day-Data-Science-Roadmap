# Day 6 — Probability Basics for Data Science

**Analogy used throughout:** Football matches 🏟️

Covers: basic probability, complement rule, independent/dependent events, joint probability, mutually exclusive events, conditional probability.

## 1. Building our football match dataset

```python
results = ["Win", "Win", "Draw", "Loss", "Win", "Win", "Draw", "Loss", "Win", "Win"]
print("Last 10 match results:", results)
print("Total matches played:", len(results))
```

## 2. Basic Probability — P(Win)

`P(A) = favorable outcomes / total outcomes`

```python
wins = results.count("Win")
total = len(results)

p_win = wins / total
print("Number of wins:", wins)
print("P(Win) =", p_win)
```

## 3. Complement Rule — P(NOT Win)

`P(A') = 1 - P(A)`

⚠️ Note: this gives P(Draw OR Loss) combined, not P(Loss) alone.

```python
p_not_win = 1 - p_win
print("P(NOT Win) =", p_not_win)

draws_losses = results.count("Draw") + results.count("Loss")
print("Check via counting:", draws_losses / total)
```

## 4. Joint Probability (Independent Events) — P(Win AND Rain)

`P(A and B) = P(A) × P(B)` — only valid when A and B are independent.

```python
p_rain = 0.3

p_win_and_rain = p_win * p_rain
print("P(Win) =", p_win)
print("P(Rain) =", p_rain)
print("P(Win AND Rain) =", p_win_and_rain)
```

## 5. Conditional Probability — P(Win | Home Game)

`P(A | B) = P(A and B) / P(B)`

```python
venues = ["Home", "Away", "Home", "Away", "Home", "Home", "Away", "Away", "Home", "Home"]

matches = list(zip(results, venues))
print("Matches with venue:", matches)

home_matches = [r for r, v in matches if v == "Home"]
print("Home matches:", home_matches)

home_wins = home_matches.count("Win")
p_win_given_home = home_wins / len(home_matches)
print("P(Win | Home) =", p_win_given_home)
```

## 6. Practice Round — Bigger Dataset (20 matches)

P(Win), P(NOT Win), and a true breakdown of P(Draw) and P(Loss) separately.

```python
new_matches = ["Win","Draw","Draw","Draw","Win","Win","Win","Loss","Loss","Loss",
               "Draw","Draw","Win","Win","Win","Win","Win","Win","Win","Win"]

new_wins = new_matches.count("Win")
new_total = len(new_matches)
new_win_ratio = new_wins / new_total
print("P(WIN) : ", new_win_ratio)

new_draws_or_losses = 1 - new_win_ratio
print("P(NOT WIN) -- DRAW OR LOSS COMBINED : ", new_draws_or_losses)

new_pure_losses = new_matches.count("Loss") / new_total
print("P(LOSS) ALONE : ", new_pure_losses)

new_pure_draws = new_matches.count("Draw") / new_total
print("P(DRAW) ALONE : ", new_pure_draws)
```

## 7. Practice Round — Conditional Probability on Bigger Dataset

```python
venues2 = ["Home", "Away", "Home", "Away", "Home", "Home", "Away", "Away", "Home", "Home",
           "Home", "Away", "Home", "Away", "Home", "Home", "Away", "Away", "Home", "Home"]

venue_matches = list(zip(new_matches, venues2))
home_matches2 = [nm for nm, vn in venue_matches if vn == "Home"]

home_wins2 = home_matches2.count("Win")
p_win_given_home2 = home_wins2 / len(home_matches2)
print("P(WIN | HOME) : ", p_win_given_home2)
```

## 8. Quiz Q1 & Q2 — Full Worked Solution

**Q1:** P(Win) and P(NOT Win) for 9 Win / 4 Draw / 2 Loss out of 15 matches.

**Q2:** P(Win AND Clean Sheet) assuming independence, P(Win)=0.6, P(Clean Sheet)=0.5.

```python
team = ["Win","Win","Win","Win","Win","Win","Win","Win","Win","Draw","Draw","Draw","Draw","Loss","Loss"]

team_win = team.count("Win")
team_total = len(team)
total_team_wins = team_win / team_total
print("P(WIN) : ", total_team_wins)

team_not_win = 1 - total_team_wins
print("P(NOT WIN) : ", team_not_win)

team_draw = team.count("Draw") / team_total
print("P(DRAW) alone : ", team_draw)

team_loss = team.count("Loss") / team_total
print("P(LOSS) alone : ", team_loss)

# Q2 -- independence assumption
p_win = total_team_wins
p_clean_sheet = 0.5
p_win_and_clean = p_win * p_clean_sheet
print("\nP(WIN AND CLEAN SHEET) -- independence assumption : ", p_win_and_clean)

# Bonus: checking against real data
Clean_sheet = ["Yes","Yes","Yes","Yes","Yes","No","No","No","No","No","Yes","No","Yes","No","Yes"]
Check_Clean_Sheet = list(zip(team, Clean_sheet))
Win_matches = [cs for tt, cs in Check_Clean_Sheet if tt == "Win"]
clean_sheets_in_wins = Win_matches.count("Yes")
p_clean_given_win = clean_sheets_in_wins / len(Win_matches)
print("P(CLEAN SHEET | WIN) -- from real data : ", p_clean_given_win)
```

## 9. Quiz Q4 — Fixed Version (P(Win | Away))

Original attempt had two bugs: mismatched list lengths in `zip()`, and looping over the raw list instead of the zipped pairs. Fixed below.

```python
r_matches = ["Win","Win","Win","Win","Win","Loss","Loss","Loss","Loss","Loss",
             "Win","Win","Win","Win","Win","Loss","Loss","Loss","Loss","Loss"]

Possibilty = ["Away","Away","Away","Away","Away","Home","Home","Home","Home","Home",
              "Away","Home","Away","Home","Away","Home","Away","Home","Away","Home"]

r_possibility = list(zip(r_matches, Possibilty))

r_away_results = [rm for rm, p in r_possibility if p == "Away"]
print("Results of away matches:", r_away_results)

away_wins = r_away_results.count("Win")
p_win_given_away = away_wins / len(r_away_results)
print("P(WIN | AWAY) : ", p_win_given_away)
```

## 🔁 Key Takeaways / Common Mistakes Log

1. Treating dependent events as independent (can't always multiply).
2. Confusing *independent* with *mutually exclusive* — different concepts.
3. Mislabeling the complement: `1 - P(Win)` ≠ P(Loss) alone — it's P(Draw OR Loss).
4. Mixing up `P(A|B)` and `P(B|A)` — direction matters, sets up Bayes' Theorem later.

**Revisit flag:** P(A\|B) vs P(B\|A) direction confusion — came up 3x in practice today.
