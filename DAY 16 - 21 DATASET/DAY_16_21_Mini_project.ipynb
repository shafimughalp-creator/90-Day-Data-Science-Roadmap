Part 1 — Day 16: Lists → Arrays, Shape, Indexing

```python
import numpy as np
```

```python
players = ["Ali", "Bilal", "Hamza", "Zain", "Sana", "Usman"]
goals_list = [9, 4, 14, 2, 7, 11]
goals = np.array(goals_list)
print(goals)
```

```python
squad_stats = np.array([
    [9, 4, 22],[4, 8, 15],[14, 3, 30],[2, 9, 10],[7, 6, 18],[11, 2, 25]
])
print(squad_stats)
```

```python
squad_stats[2, :]     # Hamza's row (index 2)
```

```python
squad_stats[:, 1]     # everyone's assists
```

Part 2 — Day 17: Broadcasting, Axis, Reshape/Transpose

```python
stat_bonus = np.array([1, 0, 2])   # +1 goal, +0 assist, +2 shots
final_stats = squad_stats + stat_bonus
print(final_stats)
```

```python
total_per_stat = squad_stats.sum(axis=0)
print(total_per_stat)
```

```python
total_per_player = squad_stats.sum(axis=1)
print(total_per_player)
```

```python
goals.reshape(-1, 1)     # (6,) -> (6,1)
```

```python
squad_stats.T             # (6,3) -> (3,6)
```

Part 3 — Day 18: Dot Product, RMSE, Cosine Similarity


```python
weights = np.array([3, 2, 1])   # goal=3, assist=2, shot=1
performance_score = np.dot(squad_stats, weights)
print(performance_score)
```

```python
# RMSE — how far off was the coach's pre-season prediction?
predicted_goals = np.array([8, 5, 12, 3, 6, 10])   # coach's pre-season guess for each player
actual_goals = goals                                # goals is already defined from Part 1 = [9, 4, 14, 2, 7, 11]

errors = actual_goals - predicted_goals
rmse = np.sqrt(np.mean(errors ** 2))

print("Errors per player:", errors)
print("RMSE (prediction accuracy):", round(rmse, 2))
```

```python
# Cosine similarity — who plays most like Hamza?
hamza = squad_stats[2]   # Hamza's row: [14, 3, 30]

for name, row in zip(players, squad_stats):
    dot = np.dot(hamza, row)
    mag_hamza = np.sqrt(np.sum(hamza**2))
    mag_row = np.sqrt(np.sum(row**2))
    similarity = dot / (mag_hamza * mag_row)
    print(f"Hamza vs {name:6s}: {similarity:.3f}")
```

Part 4 — Day 19: Boolean Masking, np.where, Random Module


```python
elite_mask = goals > 8
elite_players = np.array(players)[elite_mask]
print(elite_players)
```

```python
labels = np.where(performance_score >= 40, "Star", "Regular")
print(labels)
```

```python
np.random.seed(42)
trial_goals = np.random.randint(0, 4, size=5)
print(trial_goals)
```

Part 5 — Day 20: Series & DataFrame


```python
import pandas as pd
```

```python
goals_series = pd.Series(goals, index=players)
print(goals_series)
```

```python
squad_df = pd.DataFrame({
    "player": ["Ali","Bilal","Hamza","Zain","Sana","Usman","Zain"],
    "position": ["Forward","Midfielder","Forward",np.nan,"Defender","Forward","Defender"],
    "age": [24,27,np.nan,21,22,np.nan,21],
    "goals": [9,4,14,2,7,11,2],
    "assists": [4,8,3,9,6,2,9],
    "shots": [22,15,30,10,18,25,10]
})
print(squad_df)
```

```python
squad_df.loc[squad_df["position"] == "Forward"]
```

```python
squad_df.iloc[0:3]
```

Part 6 — Day 21: Missing Values & Duplicates


```python
squad_df.isnull().sum()
```

```python
squad_df["age"] = squad_df["age"].fillna(squad_df["age"].median())
print(squad_df["age"])
```

```python
squad_df["position"] = squad_df["position"].fillna(squad_df["position"].mode()[0])
print(squad_df["position"])
```

```python
squad_df.duplicated()
```

```python
squad_df.drop_duplicates(subset=["player"], keep="first")
```

```python
# weights was defined earlier in Part 3:
weights = np.array([3, 2, 1])   # goal=3pts, assist=2pts, shot=1pt

# clean_stats — pull the 3 numeric columns back out of squad_df as a plain array
clean_stats = squad_df[["goals", "assists", "shots"]].values

squad_df["performance_score"] = np.dot(clean_stats, weights)
squad_df["status"] = np.where(squad_df["performance_score"] >= 40, "Star", "Regular")
final_report = squad_df.sort_values("performance_score", ascending=False).reset_index(drop=True)

print(final_report)
```
