# Day 2 — Measures of Central Tendency

Mean, Median, and Mode — and when each one lies to you.

**Roadmap reference:** 90-Day Data Science Roadmap, Day 2 (Phase 1: Statistics Foundations)


## Setup

```
import numpy as np
from scipy import stats

# Sample salary dataset (with one outlier)
salaries = np.array([42000, 45000, 47000, 48000, 50000, 51000, 53000, 300000])
salaries
```

## 1. Mean

The average. Sensitive to outliers — every value pulls on it equally,
but extreme values pull harder in absolute terms.

```
mean_salary = np.mean(salaries)
print("Mean salary:", mean_salary)
```

## 2. Median

The middle value once sorted. Only cares about position, not magnitude —
this is why it's resistant to outliers.

```
median_salary = np.median(salaries)
print("Median salary:", median_salary)
```

## 3. Mode

The most frequent value. Works on categorical data too, unlike mean/median.

```
data = np.array([1, 2, 2, 3, 4, 2, 5])
mode_result = stats.mode(data, keepdims=True)
print("Mode:", mode_result)
```

## Practice Task

Compute all three measures on the salary dataset. Which one is the most
"honest" description of a typical salary?

```
print("Mean:  ", np.mean(salaries))
print("Median:", np.median(salaries))
print("Mode:  ", stats.mode(salaries, keepdims=True))

# Answer: Median — it isn't dragged up by the 300,000 outlier the way mean is.
```

## Quiz

1. Exam scores `[55, 58, 60, 62, 65, 98]` — mean or median better represents
   the typical score? Why?
2. Categories `["red","red","blue","green","red","blue"]` — which measure
   applies, and what's the answer?
3. True/False: in a perfectly symmetric distribution, mean = median = mode.

```
scores = np.array([55, 58, 60, 62, 65, 98])
print("Mean:  ", np.mean(scores))
print("Median:", np.median(scores))
```

## Key Takeaway

> If mean and median are far apart, trust the median — it's resistant to
> outliers. Mode is your only option for categorical data.
