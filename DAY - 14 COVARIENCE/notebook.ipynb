## Theory

**Covariance** tells you whether two variables move together (positive), move opposite (negative), or show no consistent pattern (near zero). Its raw size is unreliable on its own because it depends on the units of measurement.

**Correlation** rescales covariance by dividing by the standard deviations of both variables:

```
Corr(X, Y) = Cov(X, Y) / (std(X) * std(Y))
```

This produces a unit-free number always between -1 and +1, making it directly comparable across any pair of variables.

## Example 1 — Building the Dataset

````python
# Synthetic FC Lahore Lions dataset (seed=42 for reproducibility)
n_players = 25

df = pd.DataFrame({
    "minutes_played": np.random.randint(200, 900, n_players),
    "passes_completed": np.random.randint(150, 600, n_players),
    "possession_pct": np.random.uniform(35, 65, n_players),
    "goals_conceded": np.random.randint(0, 15, n_players)
})

# distance_km depends on minutes_played (plus noise) so we get a real relationship to see
df["distance_km"] = df["minutes_played"] * 0.011 + np.random.normal(0, 1, n_players)

features = df[["minutes_played", "distance_km", "passes_completed",
                "possession_pct", "goals_conceded"]]

features.head()
````

## Example 2 — Covariance & Correlation Matrices

````python
# Covariance matrix — direction only, values not directly comparable across pairs
cov_matrix = features.cov()
cov_matrix
````

````python
# Correlation matrix — unit-free, always between -1 and +1, directly comparable
corr_matrix = features.corr()
corr_matrix
````

## Summary

- Covariance shows direction of a relationship but is unreliable in raw size due to units.
- Correlation rescales covariance to a unit-free -1 to +1 range, making it comparable across any pair of variables.
- The diagonal of a correlation matrix is always exactly 1.0.
- Correlation only detects linear relationships — a value near 0 doesn't rule out more complex patterns.
- These matrices form the mathematical foundation for PCA, coming up in Phase 4.
