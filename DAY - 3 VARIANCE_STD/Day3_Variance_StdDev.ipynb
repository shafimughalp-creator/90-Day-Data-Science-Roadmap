# Day 3 — Measures of Spread: Variance & Standard Deviation

**90-Day Data Science Roadmap**

Two datasets can have the **same mean** but be completely different in reality. This notebook explores Range, Variance, and Standard Deviation — and the famous numpy vs pandas `ddof` trap.

## Setup

```python
import numpy as np
import pandas as pd

dept_a = np.array([50, 51, 49, 50, 52, 48, 50])   # Dept A salaries (in 'k)
dept_b = np.array([30, 70, 20, 80, 35, 65, 40])   # Dept B salaries (in 'k)
```

## 1. Mean — looks similar, doesn't tell the full story

```python
print("Dept A mean:", dept_a.mean())
print("Dept B mean:", dept_b.mean())
```

## 2. Range — quick, but fragile to outliers

```python
print("Dept A range:", dept_a.max() - dept_a.min())
print("Dept B range:", dept_b.max() - dept_b.min())
```

## 3. Variance — average squared deviation from the mean

```python
print("Dept A variance (numpy, ddof=0):", np.var(dept_a))
print("Dept B variance (numpy, ddof=0):", np.var(dept_b))
```

## 4. Standard Deviation — the human-readable version

```python
print("Dept A std (numpy, ddof=0):", np.std(dept_a))
print("Dept B std (numpy, ddof=0):", np.std(dept_b))
```

**Conclusion:** Dept A's salaries deviate by ~1.2k on average — very consistent. Dept B's deviate by ~21.2k — much more scattered. Same mean, very different reality.

## 5. The ddof Trap — numpy vs pandas

- `numpy.var()` defaults to `ddof=0` -> **population** variance (divide by n)
- `pandas.var()` defaults to `ddof=1` -> **sample** variance (divide by n-1, Bessel's correction)

Same data, two different (both correct, but different) answers.

```python
s = pd.Series(dept_a)
print("Pandas var (ddof=1 default):", s.var())
print("Numpy var (ddof=0 default):", np.var(dept_a))

# Make them match explicitly:
print("\nMatched -> numpy ddof=1:", np.var(dept_a, ddof=1))
print("Matched -> pandas ddof=0:", s.var(ddof=0))
```

## 6. Common Mistakes — Demonstrated

```python
# Mistake: trusting range when there's an outlier
salaries = np.array([48, 49, 50, 51, 500])
print("Range (misleading):", salaries.max() - salaries.min())
print("Std Dev (more honest):", np.std(salaries))
```

## 7. Edge Cases

```python
# All identical values -> variance and std are both 0
identical = np.array([50, 50, 50, 50])
print("Variance:", np.var(identical))
print("Std Dev:", np.std(identical))

# Sample variance with n=1 -> division by zero -> NaN
single_value = pd.Series([50])
print("Sample variance with n=1 (ddof=1):", single_value.var())
```
