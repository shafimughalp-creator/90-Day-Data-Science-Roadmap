## 1. Setup

We use `scipy.stats` for the t-tests and `numpy` to generate reproducible
synthetic data (same `seed=42` used across the roadmap).

```
import numpy as np
from scipy import stats

# Reproducibility — same seed used throughout the roadmap
np.random.seed(42)
```

## 2. Two-sample t-test — Men's vs Women's salaries

We have TWO real, separate datasets (men's salaries, women's salaries).
Comparing two real samples against each other → **two-sample t-test**.

```
# Simulate monthly salaries (in '000 PKR) for two independent groups
men_salaries = np.random.normal(loc=250, scale=30, size=20)    # mean=250k, std=30k
women_salaries = np.random.normal(loc=230, scale=28, size=20)  # mean=230k, std=28k

print("Men's avg salary:", round(men_salaries.mean(), 2))
print("Women's avg salary:", round(women_salaries.mean(), 2))
```

```
# Run the two-sample independent t-test
t_stat, p_value = stats.ttest_ind(men_salaries, women_salaries)

print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05
if p_value < alpha:
    print("Result: p < 0.05 -> Reject H0 -> difference IS statistically significant")
else:
    print("Result: p >= 0.05 -> Fail to reject H0 -> not enough evidence of a difference")
```

**Interpretation:** p = 0.016 is below α = 0.05, so we reject H0. There is
statistically significant evidence of a salary difference between the two
groups in this sample. (Not "proven" — just strong evidence.)

## 3. One-sample t-test — Men's salary vs known league average

Here we compare ONE dataset (men's salaries) against a SINGLE known fixed
value (league average = 235k) — not a second dataset.
This is exactly when you use a **one-sample t-test**.

```
# Known fixed reference value (not a dataset)
league_avg = 235

t_stat_1, p_value_1 = stats.ttest_1samp(men_salaries, league_avg)

print("T-statistic:", round(t_stat_1, 4))
print("P-value:", round(p_value_1, 4))

if p_value_1 < alpha:
    print("Result: p < 0.05 -> Reject H0 -> men's avg salary differs from league avg")
else:
    print("Result: p >= 0.05 -> Fail to reject H0 -> not enough evidence men's avg differs from league")
```

**Interpretation:** p = 0.1422 is above α = 0.05, so we fail to reject H0.
Even though the sample average (244.86k) looked different from 235k, there
isn't enough statistical evidence to call that a real difference rather than
random sampling noise.

## 4. Bonus example — Home vs Away match scores (two-sample)

Another two-sample case: two real datasets (home scores, away scores).

```
home_scores = np.random.normal(loc=2.1, scale=0.8, size=25)
away_scores = np.random.normal(loc=1.6, scale=0.7, size=25)

t_value, p_test = stats.ttest_ind(home_scores, away_scores)

print("T-statistic:", round(t_value, 4))
print("P-value:", round(p_test, 4))

if p_test < alpha:
    print("Result: p < 0.05 -> Reject H0 -> significant difference between home and away scores")
else:
    print("Result: p >= 0.05 -> Fail to reject H0 -> not enough evidence of a difference")
```

## 5. Key Takeaways — Day 10

| Test | When to use |
|---|---|
| **One-sample t-test** (`ttest_1samp`) | Comparing ONE sample's mean against a SINGLE known fixed value |
| **Two-sample t-test** (`ttest_ind`) | Comparing the means of TWO real, independent datasets |

**Decision rule:** if p-value < α (0.05) → reject H0 (statistically significant).
If p-value ≥ α → fail to reject H0 (not enough evidence).

**Correct wording:** never say a result is "proven." Say "strong evidence
against H0" or "statistically significant difference."
