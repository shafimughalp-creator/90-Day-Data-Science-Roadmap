# Day 12 — Central Limit Theorem (CLT)

**Topic:** Central Limit Theorem — Why Normal Distributions Appear Everywhere

**Hook:** No matter how messy or random your raw data is, if you take enough samples and average them, you always get a bell curve. Demonstrated below using a fake FC Lahore Lions dataset (shots-on-target per match).

## Piece 1: Setup + build a messy uniform population

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # FC Lahore Lions seed, keeps results repeatable

# Pretend: shots-on-target for Lahore Lions across 10,000 imaginary matches
# Uniform distribution = totally flat, every value between 0 and 20 equally likely
population = np.random.uniform(low=0, high=20, size=10000)

print("Population mean:", population.mean())
print("Population std:", population.std())
```

**Plain Walkthrough:** Build 10,000 fake matches with shots-on-target uniformly random between 0-20. Print the true average and spread.

## Piece 2: Prove the population itself is NOT bell-shaped

```python
plt.figure(figsize=(6,4))
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.title("Population Distribution (Uniform) - NOT a bell curve")
plt.xlabel("Shots on Target")
plt.ylabel("Frequency")
plt.show()
```

**Plain Walkthrough:** Histogram of the raw population data. Should look flat/rectangular - no bell shape.

## Piece 3: The actual CLT simulation - sample 1000 times, average each time

```python
sample_means = []

for i in range(1000):
    sample = np.random.choice(population, size=30, replace=False)  # grab 30 "matches"
    sample_means.append(sample.mean())  # record the average of that sample

sample_means = np.array(sample_means)

print("Mean of sample means:", sample_means.mean())
print("Std of sample means (this IS the Standard Error):", sample_means.std())
```

**Plain Walkthrough:** Loop 1000 times, each time randomly pulling 30 matches and recording their average. This builds the sampling distribution.

## Piece 4: Plot the sample means - watch the bell curve appear

```python
plt.figure(figsize=(6,4))
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title("Sampling Distribution of the Mean - CLT in action!")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

**Plain Walkthrough:** Histogram of the 1000 sample averages. Should now look like a clean, symmetric bell curve centered near 10.

## Piece 5: Verify the Standard Error formula matches your simulation

```python
theoretical_SE = population.std() / np.sqrt(30)
actual_SE = sample_means.std()

print("Theoretical SE (sigma/sqrt(n)):", theoretical_SE)
print("Actual SE (from simulation):", actual_SE)
```

**Plain Walkthrough:** Theoretical SE uses the population's std divided by sqrt(30). Actual SE comes from the simulation. Both should be close - proving the formula predicts reality.

## Piece 6: Prove it works on a DIFFERENT shaped distribution (exponential)

```python
population_exp = np.random.exponential(scale=5, size=10000)  # skewed, not flat

sample_means_exp = []
for i in range(1000):
    sample = np.random.choice(population_exp, size=30, replace=False)
    sample_means_exp.append(sample.mean())

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(population_exp, bins=30, color='salmon', edgecolor='black')
plt.title("Population (Exponential) - heavily skewed")

plt.subplot(1,2,2)
plt.hist(sample_means_exp, bins=30, color='lightgreen', edgecolor='black')
plt.title("Sample Means - bell shape AGAIN")
plt.tight_layout()
plt.show()
```

**Plain Walkthrough:** Build a skewed exponential population (lots of small values, long tail of big ones). Repeat the same sampling loop. Left plot shows the skewed raw data, right plot shows the sample means - still a clean bell curve, proving CLT works regardless of the original shape.

## Quiz Notes / Common Mistakes Recap

1. CLT works on ANY population shape - uniform, exponential, etc. - not just already-normal data.
2. Standard deviation (sigma) describes raw data spread. Standard Error describes spread of sample MEANS. Never mix these up.
3. Use n >= 30 per sample as a rule of thumb, especially with skewed populations.
4. Theoretical SE always uses the population's std (`population.std() / sqrt(n)`), not the std of already-simulated sample means.
