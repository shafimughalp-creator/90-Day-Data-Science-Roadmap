# Day 1 — Why Statistics? The DS Foundation

Goal: understand population vs sample, variable types, and write my first NumPy arrays.

## 1. Import NumPy
Checking the library imports correctly.

```python
import numpy as np

print(np.__version__)  # confirms numpy is installed and importable
```

## 2. Numerical Data — Continuous
Height is continuous: it can take any value in a range.

```python
# heights in cm for a small sample of 6 people
heights = np.array([162.5, 170.2, 158.9, 180.1, 165.4, 174.8])

print(heights)
print(type(heights))
```

## 3. Numerical Data — Discrete
Number of siblings is discrete: only whole, countable values.
This also shows NumPy's vectorized math — adding 1 to every value at once.

```python
# number of siblings (discrete numerical data)
siblings = np.array([0, 2, 1, 3, 0, 2])

# vectorized operation - no loop needed
siblings_next_year = siblings + 1
print(siblings_next_year)
```

## 4. Categorical Data — Nominal vs Ordinal
Cities have no natural order (nominal).
Satisfaction levels do have a natural order (ordinal).

```python
cities = np.array(["Lahore", "Karachi", "Multan", "Lahore", "Quetta"])
satisfaction = np.array(["Low", "High", "Medium", "Low", "High"])

print(cities)
print(satisfaction)
```

## 5. First Taste of Descriptive Statistics
Using NumPy to summarize the data I just created.

```python
print("Mean height:", np.mean(heights))
print("Max siblings:", np.max(siblings))
print("Unique cities:", np.unique(cities))
```

## Key Takeaways
- A sample must be **representative**, not just large, to avoid sampling bias.
- Variable type (numerical vs categorical, continuous/discrete, nominal/ordinal) decides which statistics and charts are valid.
- NumPy arrays support vectorized math; plain Python lists do not.
