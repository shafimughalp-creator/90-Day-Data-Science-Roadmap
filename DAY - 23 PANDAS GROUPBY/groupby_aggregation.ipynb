# Pandas — GroupBy & Aggregation

## Objective
Learn how to split a DataFrame into groups and compute summary statistics per group using `groupby()` and `.agg()` — single aggregations, multi-column/multi-function aggregations, and named aggregations.

```python
# Imports
import pandas as pd
import numpy as np
```

## Theory
`groupby()` works in 3 stages, always in this order:

1. **Split** — divide rows into groups based on a column
2. **Apply** — run a function on each group separately (mean, sum, count...)
3. **Combine** — merge results back into one clean table

`.agg()` lets you control exactly which function runs on which column, including naming the output columns directly (named aggregation).

```python
# FC Lahore Lions squad dataset
np.random.seed(42)  # fixes randomness for reproducibility

data = {
    "player": ["Ali", "Hamza", "Bilal", "Usman", "Zain", "Faizan", "Adeel", "Rehan"],
    "position": ["Forward", "Forward", "Midfielder", "Midfielder", "Defender", "Defender", "Forward", "Midfielder"],
    "goals": [14, 9, 3, 5, 0, 1, 11, 2],
    "rating": [8.1, 7.4, 7.0, 6.8, 6.5, 6.2, 7.9, 6.6]
}

df = pd.DataFrame(data)
df
```

## Example 1
**Single-column groupby + single aggregation.**

`df.groupby(column)[target_column].aggregation_function()`

```python
avg_goals_by_position = df.groupby("position")["goals"].mean()
# groupby("position")  -> split rows into groups by unique values in "position"
# ["goals"]            -> select only the "goals" column from each group
# .mean()              -> apply: average within each group

print(avg_goals_by_position)
```

## Example 2
**Multi-column, multi-function aggregation with `.agg()`.**

`df.groupby(col).agg({col1: "func1", col2: "func2"})`

```python
squad_summary = df.groupby("position").agg({"goals": "sum", "rating": "mean"})
# groupby("position")   -> split into groups by position
# .agg({...})           -> apply a DIFFERENT function to DIFFERENT columns, in one call
# "goals": "sum"        -> key=column name, value=function to apply to that column
# "rating": "mean"      -> another column, another function

print(squad_summary)
```

**Named aggregation** — gives self-explanatory output column names directly.

`.agg(new_name=(source_column, function))`

```python
squad_named = df.groupby("position").agg(
    total_goals=("goals", "sum"),      # new_col_name = (source_column, function)
    avg_rating=("rating", "mean")      # new_col_name = (source_column, function)
)

print(squad_named)
```

## Mini Challenge
Build a single DataFrame called `squad_report` using named aggregation that shows, per `position`:
- `total_goals` (sum of goals)
- `avg_rating` (mean of rating)
- `top_scorer_goals` (max of goals)
- `squad_size` (count of players)

```python
squad_report = df.groupby("position").agg(
    total_goals=("goals", "sum"),
    avg_rating=("rating", "mean"),
    top_scorer_goals=("goals", "max"),
    squad_size=("player", "count")
)

print(squad_report)
```

## Summary
- `groupby()` splits rows into groups; it needs a column selection or `.agg()` to actually compute anything.
- `.agg({col: func})` applies different functions to different columns in a single call.
- Named aggregation (`name=(col, func)`) produces clean, report-ready column names — the professional standard.
- `"count"` counts rows per group regardless of the underlying values.
- This is the exact Pandas mirror of SQL's `GROUP BY` + aggregate functions, coming up again in Phase 3.
