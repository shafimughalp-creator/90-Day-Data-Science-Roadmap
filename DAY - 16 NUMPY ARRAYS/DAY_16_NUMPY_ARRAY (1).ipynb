 Indexing and slicing (1D and 2D)

```python
goals_array = np.array([5, 8, 2, 12, 6])

print(goals_array[0])      # first player's goals
print(goals_array[1:4])    # players 2 to 4 (index 1,2,3)
print(goals_array[-1])     # last player's goals
```

```python
lions_goals = np.array([
    [1, 0, 2],   # Player A
    [0, 1, 1],   # Player B
    [2, 2, 0]    # Player C
])

print(lions_goals[0, 2])    # Player A, match 3 → single value
print(lions_goals[:, 1])    # everyone's goals in match 2 → a column
print(lions_goals[1, :])    # Player B's goals in all matches → a row
```
