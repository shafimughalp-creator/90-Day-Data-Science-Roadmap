**Today's goal:** understand how to update a belief using new evidence — and see why a "99% accurate" test can still mislead you when the event is rare.

## The Formula

Derived from conditional probability:

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

| Term | Symbol | Meaning |
|---|---|---|
| Prior | P(A) | Belief before evidence |
| Likelihood | P(B\|A) | How likely the evidence is, IF A is true |
| Evidence | P(B) | Overall likelihood of the evidence |
| Posterior | P(A\|B) | Belief after evidence |

## Example 1 — Spam Filter

Question: an email contains the word "FREE". What's the probability it's spam?

```python
# Prior: what fraction of all emails are spam, historically
p_spam = 0.20          # P(Spam) = 20% of emails are spam

# Likelihood: probability the word "FREE" appears, GIVEN the email type
p_free_given_spam = 0.80       # P("FREE" | Spam)
p_free_given_not_spam = 0.10   # P("FREE" | Not Spam)

# Prior for "not spam" (complement rule)
p_not_spam = 1 - p_spam
```

```python
# Evidence: P("FREE") = P("FREE"|Spam)*P(Spam) + P("FREE"|NotSpam)*P(NotSpam)
p_free = (p_free_given_spam * p_spam) + (p_free_given_not_spam * p_not_spam)

print(f"P(FREE) overall = {p_free}")
# Expected output: P(FREE) overall = 0.24
```

```python
# Bayes' Theorem: P(Spam | FREE) = P(FREE|Spam) * P(Spam) / P(FREE)
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P(Spam | email contains 'FREE') = {p_spam_given_free:.2f}")
# Expected output: P(Spam | email contains 'FREE') = 0.67
```

## Example 2 — The Rare Disease Paradox

Disease prevalence: 1 in 10,000. Test is 99% accurate both ways. Patient tests positive — how worried should they be?

```python
p_disease = 1/10000          # Prior: very rare
p_no_disease = 1 - p_disease

p_pos_given_disease = 0.99       # Test catches it 99% of the time
p_pos_given_no_disease = 0.01    # False positive rate = 1%

p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)
p_disease_given_positive = (p_pos_given_disease * p_disease) / p_positive

print(f"P(Disease | Positive Test) = {p_disease_given_positive:.4f}")
# Expected output: P(Disease | Positive Test) = 0.0098 (< 1%!)
```

**Why so low?** The disease is rare, so even a small false-positive rate applied to the huge healthy population generates more false alarms than the rare true positives. **The prior controls the result more than the test's "accuracy" does.**

Let's prove that by changing ONLY the prior — same test, disease now common (1 in 3):

```python
p_disease_common = 1/3
p_no_disease_common = 1 - p_disease_common

p_positive_common = (p_pos_given_disease * p_disease_common) + (p_pos_given_no_disease * p_no_disease_common)
p_disease_given_positive_common = (p_pos_given_disease * p_disease_common) / p_positive_common

print(f"P(Disease | Positive Test), common disease = {p_disease_given_positive_common:.4f}")
# Expected output: 0.9800 -- same test, totally different trustworthiness!
```

## Common Mistakes

1. **Confusing P(A\|B) with P(B\|A)** — the prosecutor's fallacy.
2. **Forgetting to divide by P(B)** — gives a joint probability, not a valid conditional probability.
3. **Ignoring the prior (base rate)** — "99% accurate" tells you little when the event is rare.

```python
# WRONG vs RIGHT — Mistake 1
wrong_answer = p_pos_given_disease                       # treats P(Pos|Disease) as if it equals P(Disease|Pos)
right_answer = (p_pos_given_disease * p_disease) / p_positive

print(f"Wrong (swapped): {wrong_answer}")
print(f"Right (full Bayes): {right_answer:.4f}")
```

## Connection to ML — Naive Bayes

Naive Bayes combines multiple **independent** pieces of evidence by **multiplying their likelihoods**:

$$P(Spam \mid link, free, 3AM) \propto P(link|Spam) \times P(free|Spam) \times P(3AM|Spam) \times P(Spam)$$

It's called "naive" because it assumes the signals don't influence each other — not perfectly true in real life, but works surprisingly well in practice.

## Today I Learned

- Bayes' Theorem updates a belief using new evidence: `Posterior = (Likelihood x Prior) / Evidence`
- A rare event + an imperfect test = a positive result far less trustworthy than the test's "accuracy" suggests
- Naive Bayes combines multiple independent evidences by multiplying their likelihoods together
