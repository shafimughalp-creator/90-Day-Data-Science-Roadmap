```python
staff = pd.DataFrame(
    {
        "emp_id": [101, 102, 103, 104, 105],
        "name": ["Zainab", "Omar", "Tariq", "Hira", "Kamran"],
        "dept_id": ["D1", "D2", "D1", "D3", "D2"],
        "salary": [120000, 95000, 110000, 140000, 105000],
    }
)

departments = pd.DataFrame(
    {
        "dept_id": ["D1", "D2", "D3"],
        "dept_name": ["Data Science", "Backend Engineering", "DevOps"],
    }
)
```

Task: Merge the two tables so every employee has their department name attached. Then, use .groupby() and named aggregations (.agg()) to calculate the total departmental payroll (sum) and average salary (mean) for each department name.

```python
merge_staff_dept = pd.merge(staff,departments,on="dept_id",how="outer")
print(merge_staff_dept)
```

```python
agg_of_dept = merge_staff_dept.groupby("dept_name").agg(
    total_Department_payroll=("salary","sum") ,
    average_salary=("salary","mean")
)
print(agg_of_dept)
```

```python
users = pd.DataFrame({"user_id": [1, 2, 3], "username": ["dev_shafi", "data_pro", "ml_wizard"]})

orders = pd.DataFrame(
    {"order_id": [901, 902], "user_id": [1, 1], "product_id": ["P-10", "P-20"]}  # Notice user 2 and 3 have no orders
)

products = pd.DataFrame(
    {"product_id": ["P-10", "P-20", "P-30"], "price": [4500.0, 1200.0, 8500.0]}
)
```

Merge users with orders in a way that keeps all registered users, even those without orders

Merge that intermediate result with products to attach the price column, ensuring you do not drop users who haven't ordered anything[

```python
# ✅ The Industry Standard Pipeline: Anchor to users with how="left"
step_1 = pd.merge(
    users, orders, on="user_id", how="left"
)  # keeps all users, ignores unknown orders
final_report = pd.merge(
    step_1, products, on="product_id", how="left"
)  # keeps all users, ignores unsold products (P-30 drops out!)

print(final_report)
```

```python
import pandas as pd

# Table 1: Registered loyalty customers
customers = pd.DataFrame(
    {
        "customer_id": [101, 102, 103, 104],
        "name": ["Hamza", "Ayesha", "Bilal", "Sara"],
        "loyalty_tier": ["Gold", "Silver", "Gold", "Bronze"],
    }
)

# Table 2: This week's transactions (Notice customer 104 has no transactions)
transactions = pd.DataFrame(
    {
        "txn_id": [5001, 5002, 5003, 5004],
        "customer_id": [101, 102, 101, 103],
        "item_id": ["M1", "M2", "M3", "M1"],
    }
)

# Table 3: Menu catalog (Notice M4: Cheesecake is on the menu, but nobody bought it this week)
menu = pd.DataFrame(
    {
        "item_id": ["M1", "M2", "M3", "M4"],
        "item_name": ["Caramel Macchiato", "Iced Americano", "Croissant", "Cheesecake"],
        "price": [850, 600, 450, 950],
    }
)
```

Build a multi-table merge chain that anchors to the customers table and attaches the transaction details and menu prices. You must ensure every registered customer survives the merge, while guaranteeing unsold menu items (like Cheesecake) do not leak into the table. Save this as master_df

```python
customer_trans = pd.merge(customers , transactions , on="customer_id" , how="left")
print(customer_trans)
```

```python
master_df = pd.merge(customer_trans , menu , on="item_id" , how="right")
print(master_df)
```

```python
no_purchase = master_df[master_df['customer_id'].isnull()]
print(no_purchase)
```

Using .groupby() and .agg() on master_df, generate a summary table grouped by name and loyalty_tier that calculates:

total_spent → sum of price

avg_order_value → mean of price

```python
loyalty_tier = master_df.groupby('loyalty_tier').agg(
    total_spent=("price","sum") ,
    avg_order_value=("price","mean")
)
print(loyalty_tier)
```

```python
# 1) The Master Ledger: Anchor left continuously
customer_trans = pd.merge(customers, transactions, on="customer_id", how="left")
master_df = pd.merge(
    customer_trans, menu, on="item_id", how="left"
)  # ✅ how="left" ignores Cheesecake and preserves Sara!
print(master_df)

# 2) The Inactive Audit: Now correctly isolates Sara!
no_purchase = master_df[master_df["item_id"].isnull()]
print(no_purchase)

# 3) The Revenue Report: Grouping by both name AND loyalty_tier as requested
revenue_report = master_df.groupby(["name", "loyalty_tier"]).agg(
    total_spent=("price", "sum"), avg_order_value=("price", "mean")
)
print(revenue_report)
```

```python
sales_reps = pd.DataFrame(
    {"rep_id": [1, 2, 3], "rep_name": ["John", "Fatima", "Ali"], "region_id": ["R1", "R2", "R1"]}
)

regions = pd.DataFrame({"region_id": ["R1", "R2"], "region_name": ["North", "South"]})

closed_deals = pd.DataFrame(
    {"deal_id": [101, 102, 103, 104], "rep_id": [1, 2, 1, 3], "deal_value": [5000, 8000, 12000, 4500]}
)
```

Performs an inner join chain connecting closed_deals → sales_reps → regions

```python
closed_sales = pd.merge(closed_deals , sales_reps , on="rep_id" , how="inner")
print(closed_sales)
```

```python
region_name = pd.merge(closed_sales , regions , on="region_id" , how="inner")
print(region_name)
```

Uses .agg() to calculate the total revenue (sum of deal_value) and total number of deals closed (count of deal_id) per region.

```python
total_sales = region_name.groupby("region_name").agg(
    total_revenue=("deal_value","sum") ,
    total_deals=("deal_id","count")
)
print(total_sales)
```
