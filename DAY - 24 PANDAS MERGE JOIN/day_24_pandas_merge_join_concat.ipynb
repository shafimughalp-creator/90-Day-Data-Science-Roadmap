## Imports

```python
import pandas as pd
```

## Theory
- **Merge/Join**: combine two DataFrames sideways by matching a shared key column.
- **Concat**: stack DataFrames ‚Äî new rows (`axis=0`) or new columns (`axis=1`) ‚Äî with no key matching.
- The `how` parameter (`inner`, `left`, `right`, `outer`) controls which rows survive when a match is missing.

## Example 1 ‚Äî Inner merge

```python
customers = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Ali", "Sara", "Bilal"]
})

orders = pd.DataFrame({
    "id": [1, 1, 2],
    "amount": [2500, 900, 1800]
})

# how="inner" keeps only rows where the id exists in BOTH tables
report = pd.merge(customers, orders, on="id", how="inner")
print(report)
```

## Example 2 ‚Äî Left merge + finding unmatched rows

```python
# how="left" keeps every row from the left table (customers), even with no order
left_report = pd.merge(customers, orders, on="id", how="left")
print(left_report)

# boolean mask to find customers with no matching order
no_orders = left_report[left_report["amount"].isnull()]
print(no_orders)
```

## Summary
- `pd.merge()` combines tables on a shared key column; `how` controls which unmatched rows survive.
- `pd.concat()` stacks tables with no key matching ‚Äî use `.reset_index(drop=True)` to clean up the index afterward.
- Chaining merges (table ‚Üí table ‚Üí table) is the standard pattern for building real multi-source reports.
