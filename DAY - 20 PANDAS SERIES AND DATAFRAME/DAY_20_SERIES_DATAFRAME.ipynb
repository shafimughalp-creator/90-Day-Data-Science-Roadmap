```python
import pandas as pd                                    # pd is the universal alias, always import as pd

goals = pd.Series(                                      # pd.Series(data, index=...) → build a labeled 1D array
    [12, 9, 15, 4],                                      # data: the actual values (goals scored)
    index=["Ali", "Bilal", "Hamza", "Zain"]               # index: label for each value (player names)
)

print(goals)
```

```python
squad = pd.DataFrame({                                  # pd.DataFrame(dict) → each key becomes a column
    "player": ["Ali", "Bilal", "Hamza", "Zain"],          # column 1: player names
    "goals":  [12, 9, 15, 4],                              # column 2: goals scored
    "assists": [5, 8, 3, 10]                               # column 3: assists made
})

print(squad)
```

 Sub-concept 2: Creating DataFrames (dict / NumPy array / CSV)

```python
import numpy as np                                       # need numpy for this method

stats_array = np.array([                                 # a plain 2D numpy array — no labels yet
    [12, 5],                                              # row 0: Ali   → goals, assists
    [9, 8],                                               # row 1: Bilal → goals, assists
    [15, 3],                                              # row 2: Hamza → goals, assists
    [4, 10]                                               # row 3: Zain  → goals, assists
])

squad_from_array = pd.DataFrame(                          # wrap the array in a DataFrame
    stats_array,                                          # data source: the numpy array
    columns=["goals", "assists"],                          # name the columns (array had none)
    index=["Ali", "Bilal", "Hamza", "Zain"]                 # name the rows (array had none)
)

print(squad_from_array)
```
