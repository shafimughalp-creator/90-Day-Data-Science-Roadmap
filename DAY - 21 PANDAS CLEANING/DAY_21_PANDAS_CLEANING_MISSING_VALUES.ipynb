```python
import pandas as pd
import numpy as np

data = {
    'player': ['Ahmed', 'Bilal', 'Usman', 'Hamza', 'Zain', 'Faisal'],
    'position': ['Forward', 'Midfielder', 'Defender', 'Forward', 'Goalkeeper', 'Defender'],
    'goals': [12, 5, np.nan, 8, 0, np.nan],       # np.nan = missing value marker
    'age': [24, np.nan, 27, 22, 29, 25],
    'matches_played': [30, 28, 30, np.nan, 25, 30]
}

df = pd.DataFrame(data)   # dict keys become column names, values become the column data
print(df)
```

Now let's actually find those missing spots:


```python
df.isnull().sum()      # count missing values per column
```

```python
(df.isnull().sum() / len(df)) * 100      # % missing per column
```

 Dropping Missing Values

```python
df.dropna()  # nan anywhere drop it
```

This is where thresh and subset come in — controlled cuts instead of dropping the whole squad:


```python
df.dropna(thresh=4)              # keep row if it has at least 4 non-null values
df.dropna(subset=['goals'])      # only check the 'goals' column for missing values
```

```python
df.dropna(axis=1, thresh=5)      # drop any COLUMN with fewer than 5 non-null values
```

Write one line of code on the same df that drops rows where age is missing, but ignores missing values in every other column.

 Filling Missing Values

```python
df['goals'].fillna(df['goals'].mean())
```

```python
df['age'].fillna(df['age'].median())        # median — better when outliers exist
```

```python
df['position'].fillna(df['position'].mode()[0])
```

```python
df['goals'].fillna(method='ffill')      # copies the previous row's value into the gap
```

```python
df.fillna({'goals': df['goals'].mean(), 'age': df['age'].median(), 'position': df['position'].mode()[0]})
```
